```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression

coin_gecko_list = ["ethereum", "terra-luna", "oec-binance-coin", "solana", "avalanche-2", "fantom", "matic-network",
                   "crypto-com-chain", "tron"]

column_names = ["L1_name", "Current_Price", "Price_at_double_TVL", "Price_at_triple_TVL"]
TVL_df = pd.DataFrame(columns = column_names)
TVL_list_pred = []
for crypto in coin_gecko_list:
    
    filename = f"/Users/nicolasmartinez/PycharmProjects/DefiLama/{crypto}.csv"
    #print(filename)

    dataset = pd.read_csv(filename)
    dataset.dropna()
    X = dataset.iloc[:, 2:-1].values
    y = dataset.iloc[:, -1].values
    lin_reg = LinearRegression()
    lin_reg.fit(X, y)
    plt.scatter(X, y, color = 'red')
    plt.plot(X, lin_reg.predict(X), color = 'blue')
    plt.title(f'{crypto} Price vs TVL r-squared {round(lin_reg.score(X, y),2)}')
    plt.xlabel('TVL in Millions')
    plt.ylabel('Price')
         
    plt.savefig(f'{crypto}_TVL_Price.png')
    plt.show()
    
    #getting latest TVL
    array_length = len(X)
    latest_TVL = X[array_length - 1]
    #getting latest price
    array_length = len(y)
    latest_price = y[array_length - 1]
    
    print(f"latest TVL = {lastest_TVL}")
    double_tvl =  latest_TVL*2
    triple_tvl =  latest_TVL*3
    
    price_double_TVL = lin_reg.predict([double_tvl]).item()
    price_triple_TVL = lin_reg.predict([triple_tvl]).item()
       
    chain_list = [crypto, round(latest_price,3),round(price_double_TVL,3),round(price_triple_TVL,3)]
    print(f"chain list {chain_list}")

    TVL_df.loc[len(TVL_df)] = chain_list

                
                
TVL_df.to_csv("TVL_prediction.csv")
```
